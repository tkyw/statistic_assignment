```python
def money_converter(amount:str):
    amount1, amount2 = amount.split(' - ')
    amount1, amount2 = float(amount1.strip().replace(',', '')), float(amount2.strip().replace(',', ''))
    return (amount2 + amount1) / 2


def salary_converter(item: str):
    if item == 'Full-time student' or item == 'Unemployed':
        return 0
    elif item == 'Below 2500':
        return 2500
    else:
        return money_converter(item)
```

```python
items = [salary_converter(item) for item in items.split('\n')]
```

```python
def academic_status_determination(df):
    education = '5. Choose your highest education level'
    income = '4. Choose your income level (in terms of US dollar)'
    if (df[education] == 'College Degree' or df[education] == 'Master Degree' or df[education] == 'Doctorate Degree') and df[income] != "Full-time student":
        return 'Degree Holder'
    else:
        return 'Nondegree holder'
```

```python
for row in range(df.shape[0]):
    df.loc[row, 'category'] = academic_status_determination(df.iloc[row, :])
    # print(df.iloc[row, :]['Timestamp'])
df
```
